```python
def SMS():
    global choice
    print("***Student Management System***")
    print()
    print("Enter 1 to Add")
    print("Enter 2 to Update")
    print("Enter 3 to Delete")
    print("Enter 4 to View")
    print("Enter 5 to Exit")
    print("****************************")
    print()
    choice = input("Enter your choice: ")
    #global choice
print(SMS())

while choice != "5":
    if choice == "1":
        print("You have selected Add option")
        print()
        print(SMS())
    elif choice == "2":
        print("You have selected Update option")
        print()
        print(SMS())
    elif choice == "3":
        print("You have selected Delete option")
        print()
        print(SMS())
    elif choice == "4":
        print("You have selected View option")
        print()
        print(SMS())
    else:
        print("You have selected an invalid quary")
        print()
        print(SMS())
        
```
